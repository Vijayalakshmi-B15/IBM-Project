```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

# Load the dataset
data = pd.read_csv('/content/drive/MyDrive/synthetic_customer_data_10k.csv')

# Show the first few rows to verify the data
data.head()

# Selecting the relevant columns for the model
features = data[['Age', 'Total Spend', 'Average Order Value']]  # Adjusted column names

# Normalize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(features)

# Check the scaled data
scaled_data[:5]
```

```python
import tensorflow as tf
from tensorflow.keras import layers
from sklearn.cluster import KMeans

# Assuming `scaled_data` is already defined as your preprocessed data

input_dim = scaled_data.shape[1]  # Number of features

autoencoder = tf.keras.Sequential([
    layers.InputLayer(input_shape=(input_dim,)),  # Correct input shape
    layers.Dense(128, activation='relu'),
    layers.Dense(64, activation='relu'),
    layers.Dense(32, activation='relu'),
    layers.Dense(64, activation='relu'),
    layers.Dense(128, activation='relu'),
    layers.Dense(input_dim)
])

# Compile the model
autoencoder.compile(optimizer='adam', loss='mse')

# Train the model
autoencoder.fit(scaled_data, scaled_data, epochs=50, batch_size=32, validation_split=0.2)

# Make predictions
encoded_data = autoencoder.predict(scaled_data)

# KMeans clustering
kmeans = KMeans(n_clusters=3)  # Adjust number of clusters based on your use case
kmeans.fit(encoded_data)
labels = kmeans.labels_

# Check the first few labels
print(labels[:5])
```

```python
# Save the autoencoder model
autoencoder.save("/content/autoencoder_model.h5")
```
